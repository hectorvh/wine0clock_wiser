````python
from owslib.wfs import WebFeatureService

# Base URL of the WFS service
wfs_url = "https://sgx.geodatenzentrum.de/wfs_dlm1000"

# Connect to the WFS service
wfs = WebFeatureService(url=wfs_url, version="1.1.0")
print(wfs.identification.title)
````

````python
# List all available layers
print("Available layers:")
for layer_name in list(wfs.contents):
    print(layer_name)
````

Desired layer: dlm1000:objart_43001_f

````python
# Identify available data formats
formats = wfs.getOperationByName('GetFeature').parameters['outputFormat']['values']
print("Available output formats:", formats)
````

JSON is the desired data format.

````python
# Retrieve the JSON for layer 'dlm1000:objart_43001_f'
# Use GetFeature request to get data in JSON format
layer_name = 'dlm1000:objart_43001_f'
response = wfs.getfeature(typename=layer_name, outputFormat='application/json')
````

````python
# Read the response formatted as pretty print
import json
json_data = response.read().decode('utf-8')
parsed_json = json.loads(json_data)
print(json.dumps(parsed_json, indent=2)[:500])
````

````python
# Explore the JSON structure (print all keys at the top level)
import json
data = json.loads(json_data)
print("Top-level keys in the JSON response:", data.keys())
````

````python
# Skip the geometry
# Show only the keys and values of the data associated with each shape (without the geometry)
features = data.get('features', [])
for feature in features:
    properties = feature.get('properties', {})
    print("Feature properties:")
    for key, value in properties.items():
        print(f"  {key}: {value}")
    print("-" * 40)
````

````python
# Skip the geometry
# Show only the keys "nam" and "veg" where "veg" equals 1040 for any geometry feature
# And the respective values
features = data.get('features', [])
for feature in features:
    properties = feature.get('properties', {})
    if properties.get('veg') == '1040': # veg is a string
        print(f"nam: {properties.get('nam')}, veg: {properties.get('veg')}")
````

````python
# Skip the geometry
# Find only the keys "nam" and "veg" where "veg" equals 1040 for any geometry feature
# Only print the "nam" values, not the "veg" values
# Print any "nam" value only once, even if it appears multiple times in the data
features = data.get('features', [])
unique_nam_values = set()
for feature in features:
    properties = feature.get('properties', {})
    if properties.get('veg') == '1040': # veg is a string
        nam_value = properties.get('nam')
        if nam_value not in unique_nam_values:
            print(f"nam: {nam_value}")
            unique_nam_values.add(nam_value)
````
